###    VISUALIZATION's

#  strip plot

```python
# List of categorical columns to plot against `temp`
categorical_columns = ['season', 'weekday', 'weathersit', 'holiday']

# Set up the plot grid (adjust rows and columns as needed)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Loop over the categorical columns and create strip plots
for i, col in enumerate(categorical_columns):
    ax = axes[i // 2, i % 2]  # Determine the subplot location
    sns.stripplot(x=col, y='temp', data=df1, jitter=True, palette="Set2", ax=ax)
    ax.set_title(f'Strip Plot: Temperature vs {col.capitalize()}')
    ax.set_xlabel(col.capitalize())
    ax.set_ylabel('Temperature (Celsius)')

plt.tight_layout()
plt.show()
```

```python
# This will create a set of strip plots for each of the categorical columns and show how the temperature varies across those categories.
```

# ECDF PLOT

```python
# List of categorical columns to plot ECDF for `temp`
categorical_columns = ['season', 'weekday', 'weathersit', 'holiday']

# Set up the plot grid (adjust rows and columns as needed)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Loop over the categorical columns and create ECDF plots
for i, col in enumerate(categorical_columns):
    ax = axes[i // 2, i % 2]  # Determine the subplot location
    sns.ecdfplot(data=df1, x='temp', hue=col, ax=ax, palette="Set2", legend=True)
    ax.set_title(f'ECDF Plot: Temperature vs {col.capitalize()}')
    ax.set_xlabel('Temperature (Celsius)')
    ax.set_ylabel('ECDF')

plt.tight_layout()
plt.show()
```

```python
#This will generate a set of ECDF plots showing the cumulative distribution of temperature for each category of the selected categorical columns. The ECDF curve will show the proportion of temperature values less than or equal to the corresponding temperature value along the x-axis.
```

#  HISATOGRAM PLOT

```python
numeric_columns = [ 'hr', 'holiday', 'weekday', 'weathersit', 'temp']

# Set up the plot grid
plt.figure(figsize=(12, 8))

# Iterate over numeric columns and plot histograms for each
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df1[col], bins=20, kde=True, color='skyblue', edgecolor='black')
    plt.title(f'Histogram for {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
# Skewness in a plot indicates the asymmetry of the distribution, with right skew showing a longer tail on the right and left skew showing a longer tail on the left. Clusters represent groups of closely packed data points, highlighting ranges with higher frequency or concentration of values.
```

#  PIE CHART

```python
data = {
    'instant': range(1, 611),
    'atemp': [-17.617833, -17.626278, -17.626278, -17.617833, -17.617833] * 122,
    'hum': [0.81, 0.80, 0.80, 0.75, 0.75] * 122,
    'windspeed': [0.0000, 0.0000, 0.0000, 0.0000, 0.0000] * 122,
    'casual': [3, 8, 5, 3, 0] * 122,
    'registered': [13, 32, 27, 10, 1] * 122,
    'cnt': [16, 40, 32, 13, 1] * 122,
}

# Create DataFrame
df = pd.DataFrame(data)

# Define a function to plot a pie chart for a given column
def plot_pie_chart(df2, column_name):
    # For numerical columns like casual, registered, and cnt, use value_counts
    data = df[column_name].value_counts()
    
    # Plot the pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=90, colors=['#ffcc99', '#ff6666', '#ff3333'])
    plt.title(f'Pie Chart for {column_name.capitalize()}')
    plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
    plt.show()

# List of columns for which to create pie charts
columns = ['casual', 'registered', 'cnt']

# Plot pie charts for each column
for column in columns:
    plot_pie_chart(df2, column)
```

#  Z-scores Method:

```python
import pandas as pd

# Assuming df4 is your dataset
numerical_columns = df4.select_dtypes(include=['float64', 'int64']).columns

# Create a mask to filter out the rows that contain outliers
for column in numerical_columns:
    Q1 = df4[column].quantile(0.25)
    Q3 = df4[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define the conditions for valid data (within the IQR range)
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter out outliers
    df4 = df4[(df4[column] >= lower_bound) & (df4[column] <= upper_bound)]

# Now df4 will have the outliers removed
print(df4.head(10))
```

```python
# Assuming df4 is your DataFrame and it contains numerical columns
numerical_columns = df4.select_dtypes(include=['float64', 'int64']).columns

# Define a threshold for identifying outliers using IQR (Interquartile Range)
threshold = 1.5  # The default for IQR is 1.5

# Iterate through each numerical column
for column in numerical_columns:
    # Calculate the first (Q1) and third (Q3) quartiles
    Q1 = df4[column].quantile(0.25)
    Q3 = df4[column].quantile(0.75)
    
    # Calculate IQR
    IQR = Q3 - Q1
    
    # Define the bounds for detecting outliers
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    
    # Find outliers (values outside the bounds)
    outliers = (df4[column] < lower_bound) | (df4[column] > upper_bound)
    
    # Replace outliers with the median
    median = df4[column].median()
    df4[column] = df4[column].where(~outliers, median)

# Check the first few rows of the updated DataFrame
print(df4.head())
```
